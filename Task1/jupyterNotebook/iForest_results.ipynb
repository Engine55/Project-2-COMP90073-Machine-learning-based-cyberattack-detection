```python
iforest_001_allfeat = pd.read_csv("test_result_iforest_001_allfeat.csv", encoding="utf8",low_memory = False)
iforest_001_ipca = pd.read_csv("test_result_iforest_001_ipca.csv", encoding="utf8",low_memory = False)
iforest_001_var = pd.read_csv("test_result_iforest_001_var.csv", encoding="utf8",low_memory = False)
```
