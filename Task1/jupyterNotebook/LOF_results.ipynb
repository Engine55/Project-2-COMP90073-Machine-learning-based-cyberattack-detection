```python
import pandas as pd
```

```python
lof_001_allfeat = pd.read_csv("test_result_lof_001_allfeat.csv", encoding="utf8",low_memory = False)
lof_001_ipca = pd.read_csv("test_result_lof_001_ipca.csv", encoding="utf8",low_memory = False)
lof_001_var = pd.read_csv("test_result_lof_001_var.csv", encoding="utf8",low_memory = False)
```

```python
lof_001_allfeat['conversations'] = lof_001_allfeat[['src_ip','dst_ip']].apply(lambda k: '->'.join(k),axis=1)
lof_001_allfeat.conversations.value_counts()[:15]
```

```python
lof_001_allfeat.conversations.value_counts()[:15].plot.pie()
```

```python
lof_001_ipca['conversations'] = lof_001_ipca[['src_ip','dst_ip']].apply(lambda k: '->'.join(k),axis=1)
lof_001_ipca.dst_ip.value_counts()
#.head(60)[:15]
```

```python
lof_001_ipca.conversations.value_counts()[:15].plot.pie()
```

```python
lof_001_var['conversations'] = lof_001_var[['src_ip','dst_ip']].apply(lambda k: '->'.join(k),axis=1)
lof_001_var.conversations.value_counts()[:15]
```

```python
lof_001_var.conversations.value_counts()[:15].plot.pie()
```
